```python
#pandas is a library and it contains the functions related to the database
import pandas as pd
#importing pandas for creating dataframe to perform CRUD operation
```

```python
dataset=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Regression/1.Traditional Way/1.Simple Linear Regression/Salary_Data.csv")#dataset is
#reading csv from the drive as dataset
```

```python
dataset
#Viewing the dataset
```

```python
independent=dataset[["YearsExperience"]]#
independent
#storing independent value from the dataset and viewing the independent value
```

```python
dependent=dataset[["Salary"]]
dependent
#storing dependent value from the dataset and viewing the dependent value
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
#splitting the train and test data from independent and dependent value
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
#importing linear regressor and training that model
```

```python
weight=regressor.coef_
weight
#weight is viewed
```

```python
bais=regressor.intercept_
bais
#bias is viewed
```

```python
y_pred=regressor.predict(X_test)
#prediction is made for test data
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
#evaluating model using r2score
```

```python
r_score
```
